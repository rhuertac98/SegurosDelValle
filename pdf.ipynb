```python
import boto3
from reportlab.pdfgen import canvas
from reportlab.lib.utils import ImageReader
from io import BytesIO
```

```python
# Parámetros de S3
bucket_name = "itam-analytics-werther98"
object_key = "coco/logo_SegurosDelValle.png"
local_image_path = "imagen_descargada.jpg"

# Descargar imagen desde S3
s3 = boto3.client("s3")
s3.download_file(bucket_name, object_key, local_image_path)

# Crear PDF e insertar imagen
c = canvas.Canvas("pdf_con_imagen.pdf")
c.drawString(100, 750, "Este PDF tiene una imagen desde S3")
c.drawImage(local_image_path, 100, 500, width=200, height=150)  # Ajusta posición y tamaño
c.save()

# Crear PDF en memoria
pdf_buffer = BytesIO()
c = canvas.Canvas(pdf_buffer)
c.drawString(100, 750, "PDF generado en memoria")
c.drawImage(local_image_path, 100, 500, width=200, height=150)
c.save()
pdf_buffer.seek(0)  # Volver al inicio del buffer

# Subir directamente a S3
s3 = boto3.client("s3")
bucket_name = "tu-nombre-de-bucket"
s3_path = "carpeta/pdf_con_imagen.pdf"

s3.upload_fileobj(pdf_buffer, bucket_name, s3_path)
print("PDF subido exitosamente a:", f"s3://{bucket_name}/{s3_path}")
```
